```python
import numpy as np
a = np.array([[1,2],[3,4]])
print(a)
```

```python
b = np.random.randint(1, 101, size=(3,4))
print(b)
```

```python
c = np.ones((3,3))
d = np.zeros((3,3))
print(c)
print(d)
```

```python
e = np.random.random((6,3))
print(e)

f = np.transpose(e) #invert an array rows become columns, columns become rows
print(f)
```

```python
n = 2
p = 3
g = np.random.random((n,p))
h = np.random.random((p,n))
dotprod = np.dot(g, h)
print(g)
print(h)
print(dotprod)
```
